```python
import numpy as np
import scipy
import math
```

```python
def double_factorial(n):
    res = 1
    for i in range(n, -1, -2):
        res *= i
    return res
```

```python
def compute_XZ(d1, d2, u2, x2, z2, alpha):
    res = 0
    for i in range(d1+1):
        for j in range(d2+1):
            tmp = scipy.special.binom(d1, i) * scipy.special.binom(d2, j)
            if (d1-i)%2 == 0 and (d2-j)%2 == 0 and (i+j)%2 == 0:
                u_deg = (i+j)//2
                x_deg = (d1-i)//2
                z_deg = (d2-j)//2
                
                if (i+j)==0:
                    u_moment = 1
                else:
                    u_moment = u2**(u_deg) * double_factorial(i+j-1)
                if (d1-i)==0:
                    x_moment = 1
                else:
                    x_moment = x2**(x_deg) * double_factorial(d1-i-1)
                if (d2-j)==0:
                    z_moment = 1
                else:
                    z_moment = z2**(z_deg) * double_factorial(d2-j-1)
                term = u_moment * x_moment * z_moment
            else:
                term = 0
            
            if i==j:
                tmp *= term
            elif i>j:
                alpha_term = alpha**(i-j)
                tmp *= term*alpha_term
            elif i<j:
                alpha_term = alpha**(j-i)
                tmp *= term
                tmp /= alpha_term
            # print(f"i: {i}, j:{j}, tmp:{tmp}")
            res += tmp
    return res
```

```python
d1 = 5
d2 = 7

u2 = 1
x2 = 12
z2 = 12
alpha = 1
k = 3

x2_new = x2 + (alpha**2)*u2 - k*k*(alpha**2)*u2
z2_new = z2 + (u2 / (alpha**2)) - (u2 / (alpha**2))/(k*k)
print(x2_new + u2*k*k*(alpha**2))
print(z2_new + (u2 / (alpha**2 * k * k)))
print(x2+u2*alpha*alpha)
print(z2+u2/(alpha*alpha))
assert x2_new>0, "Negative var of X"
assert z2_new>0, "Negative var of Z"

# print("Model1: ")
model1 =  compute_XZ(d1, d2, u2, x2, z2, alpha)
# print("Model2: ")
model2 =  compute_XZ(d1, d2, u2, x2_new, z2_new, k*alpha)
```
